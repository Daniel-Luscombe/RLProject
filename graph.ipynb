```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, csv
```

```python
# Read the CSV file
df = pd.read_csv('Experiments/dqn_log.csv')
x = df['Episode'].to_numpy()
y = df['Reward'].to_numpy()

plt.plot(x, y, label='Reward per Episode', color='blue', marker='o', markersize=3, linestyle='none')
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('DQN Agent Performance')
plt.legend()
plt.show()
```
